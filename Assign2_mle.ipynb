```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load and Inspect Dataset
iris = sns.load_dataset("iris")

print(iris.head())

print("\nColumns:", iris.columns)

print("\nRows and Columns:", iris.shape)
```

```python
#Class Distribution
sns.countplot(x="species", data=iris)
plt.title("Species Distribution in Iris Dataset")
plt.show()
```

```python
#Feature Distribution (Histograms)
plt.hist(iris["sepal_length"], bins=20)
plt.title("Sepal Length Distribution")
plt.show

plt.hist(iris["petal_length"], bins=20)
plt.title("Petal Length Distribution")
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
#Pair Plot
sns.pairplot(iris, hue="species")
plt.show()
```

```python
#Heatmap
corr = iris.drop(columns=["species"]).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
#Box Plot
sns.boxplot(data=iris)
plt.title("Boxplot of Iris Features")
plt.show()
```

```python
#Violin Plot
sns.violinplot(x="species", y="petal_length", data=iris)
plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
#Compare Iris vs Penguins
penguins = sns.load_dataset("penguins")

sns.scatterplot(x="flipper_length_mm", y="body_mass_g", hue="species", data=penguins)
plt.title("Penguins: Body Mass vs Flipper Length")
plt.show()
```
