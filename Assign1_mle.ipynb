```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load and Inspect Dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.to_csv("titanic_data.csv", index=False)

df = pd.read_csv("titanic_data.csv")
```

```python
print("Dataset Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nFirst 10 Rows:")
print(df.head(10))
```

```python
#Identify missing values
print("Missing Values per column")
print(df.isnull().sum())
```

```python
print("\nPercentage Missing:")
print((df.isnull().sum()/len(df))*100)
```

```python
#Clean Dataset
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
#Encode Categorical Features
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
```

```python
df['Embarked'] = df['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
df.head()
```

```python
#Summary Statistics
print(df.describe())
```

```python
print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nStandard Deviation:\n", df.std(numeric_only=True))
```

```python
print("\nSurvival Count:")
print(df['Survived'].value_counts())

print("\nSurvived:", df['Survived'].sum())
print("Did not Survive:", len(df) - df['Survived'].sum())
```

```python
#Grouped Survival Analysis
print("\nSurvival by Gender")
print(df.groupby('Sex')['Survived'].value_counts())

print("\nSurvival by Passenger Class")
print(df.groupby('Pclass')['Survived'].value_counts())

print("\nSurvival by Embarkation")
print(df.groupby('Embarked')['Survived'].value_counts())
```

```python
#Visulation
sns.countplot(x='Survived', data=df)
plt.title("Total Survivors vs Non-Survivors")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
#Histogram of Age
plt.hist(df['Age'], bins=30)
plt.title("Age Distributio of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(data=df, x="Age", hue="Survived", bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.show()
```
